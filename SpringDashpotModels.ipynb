```python
import math 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#set the parameters

k=1.25    #set the parameter k
mu=1   #damping
E=0.01   # Maxwell coefficient
beta=0.01  # stressrate coefficient
x0=1   #initial spring length
v0=0   #initial velocity
sigma0=0   #initial stress subsume mass into sigma

dt=min(0.001, 0.01*beta)
##dt= 0.001
t = np.arange(0,10,dt) #set the time parameter in steps dt

```

```python
# Flexible to look at different spring-dashpot models
def deriv(y, t, k,mu,E,beta):
    x,v,sigma = y
    dxdt = v
    dvdt = sigma
    dsigmadt = (1/beta)*(-E*sigma -(mu*v + k*x))
    return dxdt, dvdt, dsigmadt

# Initial conditions vector
y0 = x0, v0, sigma0
# Integrate the equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(k,mu,E,beta))
x,v,sigma = ret.T

##print(t,x)
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,x)
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('Dynamics of a spring-dashpot')
plt.grid(True)
```

```python

#set the parameters

k2=10    #set the parameter k
mu2=1   #damping
E2=1   # Maxwell coefficient
beta2=1  # stressrate coefficient
x0=1   #initial spring length
v0=0   #initial velocity
sigma0=0   #initial stress subsume mass into sigma

dt= 0.001
t = np.arange(0,10,dt) #set the time parameter in steps dt

# Simple spring
def deriv2(y2, t, k2,mu2):
    x2,v2 = y2
    dx2dt = v2
    dv2dt = -k2*x2 - mu2*v2

    return dx2dt, dv2dt

# Initial conditions vector
y0 = x0, v0
# Integrate the equations over the time grid, t.
ret2 = odeint(deriv2, y0, t, args=(k2,mu2))
x2,v2 = ret2.T

##print(t,x)

plt.figure(figsize=(12,6))
plt.plot(t,x2)
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('Simple spring')
plt.grid(True)
```
